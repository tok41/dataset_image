# 訓練用画像を確認する

* ファイル数が多いので、いくつかだけ表示して、どんなものか確認する
* OpenCVはBGR形式なので、RGBの画像を読ますと色がおかしい！

```python
import cv2
import sys, os
import os.path
import numpy as np
from PIL import Image

import matplotlib.pyplot as plt

%matplotlib inline
```

## 取得してきた原画像を確認する

```python
img_path = "/home/tokita/workspace/etc/get_images/raw_images/full"
files = os.listdir(img_path) 
print len(files)

plt.figure(figsize=(20, 20))
num = 10

cnt = 0
for i in np.random.permutation(len(files))[:num]:
    cnt+=1
    filename = img_path + "/" + files[i]
    #image = cv2.imread(filename)
    image = Image.open(filename)
    plt.subplot(1,num , cnt)
    plt.imshow(image)
    plt.axis('off')
```
